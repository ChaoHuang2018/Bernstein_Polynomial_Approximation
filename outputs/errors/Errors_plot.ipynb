```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
interval_errors = glob.glob("*interval_errors.txt")
lip_errors = glob.glob("*lip_errors.txt")
```

```python
interval_errors.sort()
lip_errors.sort()
```

```python
interval_errors
```

```python
lip_errors
```

```python
for (f_lip_errors, f_i_errors) in zip(lip_errors[:-1], interval_errors[:-1]):
    lip_errs = np.loadtxt(f_lip_errors)
    i_errs = np.loadtxt(f_i_errors)
    x = np.arange(lip_errs.shape[0])
    plt.figure()
    plt.plot(x, lip_errs, label = f_lip_errors[:-4])
    plt.plot(x, i_errs, label = f_i_errors[:-4])
    plt.legend()
```

```python
tora_lip_errs = np.loadtxt('nn_tora_relu_retrained_lip_errors.txt')
tora_i_errs = np.loadtxt('nn_tora_relu_retrained_interval_errors.txt')
x = np.arange(tora_lip_errs.shape[0])
plt.figure()
plt.plot(x, tora_lip_errs, label = 'nn_tora_relu_lip_errors')
plt.plot(x, tora_i_errs, label = 'nn_tora_relu_interval_errors')
plt.legend()
plt.figure()
plt.plot(x, tora_i_errs, label = 'nn_tora_relu_interval_errors', color='#ff7f0e')
plt.legend()
```
